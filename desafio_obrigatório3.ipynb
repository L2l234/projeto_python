```python
import pandas as pd
import requests
import sqlite3
```

```python
import requests

url = 'https://brasilapi.com.br/api/ibge/uf/v1'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
else:
    print("Falha ao buscar os dados da API")
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.rename(columns={'id':'ID', 'sigla': 'Sigla', 'nome': 'Nome', 'regiao': 'Regiao'})
df['Regiao'] = [regiao['nome'] for regiao in df['Regiao']]
df
```

```python
def salva_bd(df,nome_tabela):
    conn= sqlite3.connect('Estados_do_Brasil.db')
    df.to_sql(nome_tabela,conn,if_exists='replace',index=False)
    conn.close()
    return True




```
